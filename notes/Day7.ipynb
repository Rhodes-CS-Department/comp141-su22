````python
# Winner function (#2 from lab 3)
def winner(name1, votes1, name2, votes2):
    if votes1 > votes2:
        return name1
    elif votes2 > votes1:
        return name2
    else:
        return 'tie'
      
print("Winner 1: ", winner("David", 50, "Kristin", 65))
print("Winner 2: ", winner("Sally", 400, "Morgan", 400))
print("Winner 3: ", winner("Todd", 45, "Sue", 90))
````

````python
#postage function - #3 from lab-03
#This function calculates how much postage will cost, based on weight of mailing
#Parameters: ounces - a number representing weight of a piece of mail
#Returns: a floating-point number that represents the cost of mailing a letter
def postage(ounces):
    if ounces <= 0 or ounces > 13:
        return 0
    else:
        cost = .90 + (ounces-1) * .20
        return cost

print("Postage for 3 ounces: ", postage(3))
print("Postage for 0 ounces: ", postage(0))
print("Postage for 20 ounces: ", postage(20))
````

````python
# age_today, #5 from lab-03
import datetime

def age_today(month, day, year):
    today_year = datetime.date.today().year    # get today's year
    today_month = datetime.date.today().month  # get today's month as a number 1-12
    today_day = datetime.date.today().day      # get today's day of the month
  
    age = today_year - year
  
    #if birthday has yet to come this year, subtract 1 from age
    if month > today_month or (month == today_month and day > today_day):
        age = age - 1

    return age

print("Age today 9/29/2000:", age_today(9, 29, 2000))
print("Age today 7/29/2000:", age_today(7, 29, 2000))
````

````python
#8 Timezone difference function

def timezone_diff(zone1, zone2):
    #Assigns value for time1 based value of zone1  
    if zone1 == 'eastern':
        time1 = 0
    elif zone1 == 'central':
        time1 = 1
    elif zone1 == 'mountain':
        time1 = 2
    elif zone1 == 'pacific':
        time1 = 3
    #Assigns value for time2 based value of zone2  
    if zone2 == 'eastern':
        time2 = 0
    elif zone2 == 'central':
        time2 = 1
    elif zone2 == 'mountain':
        time2 = 2
    elif zone2 == 'pacific':
        time2 = 3
    
    return time1 - time2

print("Time zone difference between central and eastern: ", timezone_diff("central", "eastern"))
print("Time zone difference between eastern and central: ", timezone_diff("eastern", "central"))
print("Time zone difference between pacific and central: ", timezone_diff("pacific", "central"))
````

## While Loops

Syntax:
```
while __test__: #test is a boolean expression (evaluates to True or False)
    statement 1
    statement 2
    statement 3 
statement 4
statement 5
```
In the above example, statements 1-3 are part of the body of the loop (indented). Statements 4 and 5 will only execute after the loop body is completely done executing.

Below are flow charts for while loops (the left shows a general flow chart), the right shows the flow chart for a specific program. Let's trace through the code for the example on the right and then we can run the code below to verify our solution.



````python
#Program from right example above
number = 6
count = 0

while count <= number:
    print(count, end=', ')
    count += 1
````

## Calling Functions in a Loop
* Functions can be called from statements in the body of a loop (Often improves the design)
* Example: Write a function to calculate then display the commission for a sales amount (Call the function inside a loop)


````python
# This program calculates sales commission based on user inputs 

def calculate_commission(sales_amt, commission_rate):
    #Calculate the commission
    commission = sales_amt * commission_rate
    return commission

def main():
    keep_going = 'y'

    #Calculate a series of commissions
    while keep_going == 'y':
        sales = float(input("Enter the amount of sales: "))
        comm_rate = float(input("Enter the commission rate: "))

        commission = calculate_commission(sales, comm_rate)

        print("The commission is $", format(commission, '.2f'), '.', sep='')

        #See if the user wants to do another one.
        keep_going = input("Do you want to calculate another commission (y for yes)? ")

main()
````

## Pseudo-code
* An informal way of writing algorithms for humans to read (not computers!)
* Illustrates the logic of an algorithm, but omits details that people can fill in automatically.
* You get to make it up as you go along, as long as you (and other people) can easily understand it. 

Example: 
* Instead of saying: `name = input("What is your name? ")`, pseudocode might use a line that says `name = ask user for name`
* Instead of saying: `if x >= 0 and x <= 100: print("$%.2f" % x)`, pseudocode might say
```
if x is between 0 and 100:
	print x with 2 decimal places
```
* The point is to get your ideas down on paper quickly, so you can worry about the programming details and exact syntax later.


__To write any while loop:__
1. Write out pseudocode for what the loop does, explicitly repeating lines until you've repeated the code at least twice.
2. Include an "if" statement in your code that will be True if you want the loop to keep going.
3. Make sure the code repeats the "if" statement at least twice. 
4. Find the statements between consecutive "if" statements. These statements will become the body of the loop.
5. The "if" test will become the "while" test.
6. If there's anything before the first "if" test, it will go immediately before the while loop (outside of the body).
